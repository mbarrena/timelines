```python
import re
def detect_date(string):
    string = string.lower() #lowercase string for easier search
    
    months = ["enero", "febrero", "marzo", "abril", "mayo", "junio", "julio", "agosto", "septiembre", "octubre", "noviembre", "diciembre", "ene", "feb", "mar", "abr", "may", "jun", "jul", "ago", "set", "oct", "nov", "dic","setiembre","sep"]
    months_num = list(range(1,13))*2 + [9,9]
    day = 0
    month = 0
    year = 0
    ac = 0
    
    for i in range(len(months)): #search month name and save corresponding month number
        if(re.search(rf"\b{months[i]}\b",string)):
            month = months_num[i]
            break       
            
    rac = re.search(r"\ba\.?c\b",string) #match ac or a.c, check if year is before Christ
    ryear = re.search(r"[1-2]?[0-9]{3}",string)
    
    if(not ryear):
        #Year must be specified else the date is invalid
        raise Exception("No year found:",string)

    year=int(ryear[0])
    
    if(rac):
        ryear = re.search(r"[0-9]{2,4}",string)
        if(not ryear):
            raise Exception("No year before Christ found:",string)
        year=int("-"+ryear[0])
    
    #Find all possible days and years mentioned. If there's no manth, there will be no day. 
    rdaymonth = re.findall(r"\b[0-3]?[0-9]\b",string)
    
    #Puede pasar que el nombre del evento en si incluya numeros, intente sacar todas las ocurrencias pero los numeros de los eventos
    #siempre estan despues de la fecha asi que a lo sumo si el evento no especificaba dia o mes tendra dia o mes incorrecto
    #pero como no estaba especificado, no importa.
    
    if(len(rdaymonth)>0): #day and month may not be specified
        day = int(rdaymonth[0])
        if(len(rdaymonth)>1 and month==0):
            #Need the month check in case a month has already been detected and the extra number comes from the event name
            month = int(rdaymonth[1])

    return(year,month,day)
            

print(detect_date("1976 02 abr Plan de Martínez de Hoz (1ª etapa).pdf"))
print(detect_date("1976 16 setiembre La noche de los lápices"))
```

```python
import pandas as pd
import numpy as np
import unicodedata
IDN_SIZE = 35
def strip_accents(text):
    try:
        text = unicode(text, 'utf-8')
    except NameError: # unicode is a default on python 3 
        pass

    text = unicodedata.normalize('NFD', text)\
           .encode('ascii', 'ignore')\
           .decode("utf-8")

    return str(text)

def sanitizar_nombre(name):
    return strip_accents((name.strip().replace(" ", "_").replace("(","").replace(")","")).lower()).replace('&#34;','')

def get_name(x):
    sanitized = sanitizar_nombre(x[0])
    return sanitized[:35]

# RECIBE FILA DEL DATAFRAME
# precondición: el campo imagen no es NaN
def obtener_url_img(y):
    url = y[7].split('":"')[1]
    url2 = url.split(",")[0][:-1]
    return url2

# RECIBE FILA DEL DATAFRAME
def obtenerlink(y):
    return y[1]

def quote(str):
    return '"'+str+'"'

def unquote(str):
    if(str[0]=='"'):
        return str[1:][:-1]
    return str

# Identify type of timeline event (general information (logo_info) vs tourist information (logo_tour))
def find_type(event):
    if("HECHO TURÍSTICO" in event[0]):
        return "turismo"
    else:
        return "informacion"
```

```python
df = pd.read_csv("./links-eventos-drive.csv",index_col=False)

# First, we must identify possible repeated events
names = df.apply(get_name,axis=1)
seen = set()
check_repeats = []
for x in names:
    if x not in seen:
        seen.add(x)
    else:
        check_repeats.append(x)
handled_repeats=[]
#print(check_repeats)

def sorter(x):
    splt = x.split('start: ')[1].replace("new Date(", "").replace(")}","")
    date = splt.split(',')
    date[0]=int(date[0])
    date[1]=int(date[1])
    date[2]=int(date[2])
    if(int(date[0])<1810):
        return "1unidad"
    if(int(date[0])>=1810 and int(date[0])<1914):
        return "2unidad"
    if(int(date[0])>=1914 and int(date[0])<1930):
        return "3unidad"
    if(int(date[0])>=1930 and int(date[0])<1945):
        return "4unidad"
    if(int(date[0])>=1945 and int(date[0])<1955):
        return "5unidad"
    if(int(date[0])>=1955 and int(date[0])<1966):
        return "6unidad"
    if(int(date[0])>=1966 and int(date[0])<1976):
        return "7unidad"
    if(int(date[0])>=1976 and date<=[1983,12,10]):
        return "8unidad"
    if(int(date[0])>=1983 and date<=[1989,7,8]):
        return "9unidad"
    if(int(date[0])>=1989 and date<=[2003,5,24]):
        return "10unidad"
    if(int(date[0])>=2003 and date<=[2015,12,9]):
        return "11unidad"
    if(int(date[0])>=2015):
        return "12unidad"  
    else:
        return "otro"

start_line = "var items = new vis.DataSet(["
end_line = "]);"

print("Procesando base de datos...")
df_copy = df.copy()
events = df_copy.apply(try_create,axis=1)
events = events.dropna()
unidad = events.copy().apply(sorter)
frame = { 'Event': events, 'Unit': unidad } 
cosas = pd.DataFrame(frame) 

unit_names = ["1unidad","2unidad","3unidad","4unidad","5unidad","6unidad","7unidad","8unidad","9unidad","10unidad","11unidad","12unidad"]
units= {}
for unit in unit_names:
    units[unit] = cosas[cosas['Unit']==unit]

#display(cosas.head(20))
#raise Exception

for unit_name,unit in units.items():
    print("Guardando en archivo...")
    del unit["Unit"]
    f = open("./"+unit_name+"/"+unit_name+"_appendable.js", "w")
    f.write(start_line)
    f.write("\n")

    for _,linea in unit[:-1].iterrows():
        #lleno con todos los eventos hasta el anteultimo
        f.write(linea["Event"] + ",\n")

    #La ULTIMA fila debe ir SIN COMA en la base finalizada, y CON COMA en la appendable!
    f.close()
    print("Diferenciando versiones appendable y finalizada")
    !cp "./{unit_name}/{unit_name}_appendable.js" "./{unit_name}/{unit_name}.js"

    f2 = open("./"+unit_name+"/"+unit_name+".js","a")
    #aca va sin coma y con linea finalizadora
    f2.write(unit.iloc[-1]["Event"]+"\n")
    f2.write(end_line)
    f2.close()
    print("Version finalizada creada")

    #La version appendable va con coma y sin linea finalizadora
    f = open("./"+unit_name+"/"+unit_name+"_appendable.js", "a")
    f.write(cosas.iloc[-1]["Event"]+",")
    f.close()
    print("Versión appendable creada")
    print("Finalizado!")
```
